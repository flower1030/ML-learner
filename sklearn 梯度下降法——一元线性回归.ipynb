```python
from sklearn.linear_model import LinearRegression
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data=np.genfromtxt("Downloads/data.csv",delimiter=",") # delimiter 分隔符
```

```python
x_data=data[:,0]
y_data=data[:,1]
print(x_data)
```

```python
plt.scatter(x_data,y_data)
plt.show()
```

```python
print(x_data.shape)
```

```python
#增加X和Y的维度，好与模型格式相匹配
x_data=data[:,0,np.newaxis] #axis 轴
y_data=data[:,1,np.newaxis]
print(x_data.shape)
```

```python
#建模并训练
model=LinearRegression()
model.fit(x_data,y_data)
```

```python
plt.plot(x_data,y_data,"b.")
plt.plot(x_data,model.predict(x_data),"r")
plt.show()
```
