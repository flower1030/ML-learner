```python
#学习率
#代价函数 循环几次——
#k b
#迭代次数 在哪里迭代——偏导函数中
#偏导函数 求的是k,b 不断求出k,b带入到代价函数中
#先写出代价函数和偏导函数
```

```python
a=0.0001
n=50
k=0
b=0
```

```python
#代价函数
def computer_error(k,b,x_data,y_data):
    y=0
    for i in range(0,len(x_data)):
        y+=(y_data[i]-(k*x_data[i]+b))**2
    return y/len(x_data)/2
```

```python
#梯度下降函数
def gradient_decent_runner(k,b,x_data,y_data,a,n):
    m=len(x_data)
    for i in range(n):
        k_grad=0
        b_grad=0
        for j in range(0,len(x_data)):
            k_grad+=(k*x_data[j]+b-y_data[j])*x_data[j]*(1/m)
            b_grad+=(k*x_data[j]+b-y_data[j])*(1/m)
        k=k-a*k_grad
        b=b-a*b_grad
    return k,b
```

```python
computer_error(k,b,x_data,y_data)
```

```python
k,b=gradient_decent_runner(k,b,x_data,y_data,a,n)
print(k,b)
```
